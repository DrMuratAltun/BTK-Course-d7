```python
img_bgr=cv2.imread('CV/murataltun.png')
```

```python
img_bgr
```

```python
cv2.imshow('murat altun',img_bgr)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
b,g,r=cv2.split(img_bgr)
```

```python
cv2.imshow('mavi',b)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cv2.imshow('red',r)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cv2.imshow('yeşil',g)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_reverse=cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
cv2.imshow('BGR to RGB',img_reverse)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
new_image=cv2.merge((b,g,b))
cv2.imshow('merge',new_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
elma=cv2.imread('CV/elma.jpg')
cv2.imshow('elma',elma)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
b,g,r=cv2.split(elma)
```

```python
new_image=cv2.merge((b,r,g))
cv2.imshow('elma',elma)
cv2.imshow('yeşil',new_image)
cv2.waitKey()
cv2.destroyAllWindows()
```
